```python
import re
import numpy as np
import pandas as pd

def clean_final_grade(val):
    if pd.isna(val):
        return np.nan

    # Convert Arabic numerals to English
    val = str(val).strip().replace('Ù«', '.')
    trans = str.maketrans('Ù Ù¡Ù¢Ù£Ù¤Ù¥Ù¦Ù§Ù¨Ù©', '0123456789')
    val = val.translate(trans)

    # Handle text or weird inputs
    if re.search('[A-Za-z]', val):
        return np.nan
    if 'ÙƒØ³Ø±' in val:
        val = val.replace('ÙˆÙƒØ³Ø±', '').strip()
        try:
            return float(val) + 0.5
        except:
            return np.nan

    try:
        num = float(val)

        # Handle fractions (like 0.74)
        if 0 < num < 1:
            num *= 100

        # Handle totals (like 392 out of 410)
        elif num > 100:
            num = (num / 410) * 100

        return round(num, 2)

    except:
        return np.nan


# Apply cleaning
df['final_grade_percent'] = df['final_grade_percent'].apply(clean_final_grade)

# Fill NaN with median
median_grade = df['final_grade_percent'].median()
df['final_grade_percent'].fillna(median_grade, inplace=True)
```
