## **Step 6: Clean Study Hours (Three Columns)**
**What it does:** Cleans data about how many hours students study in different performance levels:  
- study_hours_weak - Hours studied when getting weak grades  
- study_hours_average - Hours studied when getting average grades  
- study_hours_excellent - Hours studied when getting excellent grades  

**The cleaning process:**  
- Converts Arabic numbers (ููกูขูฃ...) to English numbers (0123...)  
- Converts Arabic words to numbers:  
  - 'ุณุงุนุชูู' โ 2 hours  
  - 'ุณุช ุณุงุนุงุช' โ 6 hours  
  - 'ุณุงุนู' or 'ุณุงุนุฉ' โ 1 hour (or the number mentioned)  
- Handles ranges by taking the average:  
  - "3-5 hours" becomes 4 hours  
  - "2/4 hours" becomes 3 hours  
- Fills any remaining missing data with the median (middle value)

````python
def clean_hours(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()

    # convert Arabic digits โ English
    x = (x.replace('ู','0').replace('ูก','1').replace('ูข','2').replace('ูฃ','3')
           .replace('ูค','4').replace('ูฅ','5').replace('ูฆ','6').replace('ูง','7')
           .replace('ูจ','8').replace('ูฉ','9'))

    # translate Arabic words to numbers if no digit found
    if 'ุณุงุนุชูู' in x:
        return 2
    if 'ุณุช ุณุงุนุงุช' in x:
        return 6
    if 'ุณุงุนู' in x or 'ุณุงุนุฉ' in x:
        # if it has a number before it, use that number
        for i in range(10, 0, -1):
            if str(i) in x:
                return float(i)
        return 1  # default if no number given

    # handle ranges โ take average
    if '-' in x:
        parts = x.split('-')
        try:
            return (float(parts[0]) + float(parts[1])) / 2
        except:
            pass
    if '/' in x:
        parts = x.split('/')
        try:
            return (float(parts[0]) + float(parts[1])) / 2
        except:
            pass

    # convert plain numbers
    try:
        return float(x)
    except:
        return np.nan
````

````python
for col in ['study_hours_weak', 'study_hours_average', 'study_hours_excellent']:
    df[col] = df[col].apply(clean_hours)
    df[col] = df[col].fillna(df[col].median())
````

## **Step 9: Clean Sleep Hours (Sleep_Hours Column)**
**What it does:** Cleans how many hours teachers/students sleep per night  

**The cleaning process:**  
- Converts Arabic numbers to English  
- Removes Arabic words like 'ุณุงุนุงุช' (hours) and 'ุณุงุนู'  
- Handles ranges by taking the average:  
  - "6-8" hours becomes 7 hours

````python
def clean_sleep(x):
    x = str(x).strip()
    x = (x.replace('ู','0').replace('ูก','1').replace('ูข','2').replace('ูฃ','3')
           .replace('ูค','4').replace('ูฅ','5').replace('ูฆ','6').replace('ูง','7')
           .replace('ูจ','8').replace('ูฉ','9'))
    x = x.replace('ุณุงุนุงุช','').replace('ุณุงุนู','').strip()

    if '-' in x:
        parts = x.split('-')
        try:
            return (float(parts[0]) + float(parts[1])) / 2
        except:
            pass

    try:
        return float(x)
    except:
        return np.nan


df['sleep_hours'] = df['sleep_hours'].apply(clean_sleep)
````

## **Step 12: Clean Study Distractions (Study_Distraction Column)**
**What it does:** Takes free-form text answers about what distracts students and organizes them into categories  

**Steps:**  
- Converts text to lowercase for easier matching  
- Fixes spelling variations (e.g., 'ุงูููุจุงูู' and 'ุชูููู' both become 'ููุจุงูู')  
- Identifies categories mentioned in the answer  

**Categories created:**  
- Phone / Screens  
- Social Media / Internet  
- Friends / Social Life  
- Games / Sports  
- TV / Series  
- Family / Visits  
- Lack of Motivation / Focus  
- Other  

**Example:** If someone writes "I get distracted by my phone and YouTube," the script marks:  
- Phone / Screens: Yes  
- Social Media / Internet: Yes  

Creates new Yes/No columns for each category so you can easily analyze patterns.

````python
def clean_study_distraction(x):
    if pd.isna(x):
        return 'Other'

    text = str(x).strip().lower()

    # Normalize Arabic spellings
    text = (text.replace('ุงูุณูุดูุงู', 'ุณูุดูุงู')
                 .replace('ุงูููุฏูุง', 'ููุฏูุง')
                 .replace('ุงูููุจุงูู', 'ููุจุงูู')
                 .replace('ุชูููู', 'ููุจุงูู')
                 .replace('ุงููุงุชู ุงููุญููู', 'ูุงุชู')
                 .replace('ุงููุญููู', 'ููุจุงูู'))

    categories = []

    # Phone / Screens
    if any(word in text for word in ['ููุจุงูู', 'ูุงุชู', 'ุชููููู', 'ุดุงุดุงุช', 'ููุจููุชุฑ']):
        categories.append('Phone / Screens')

    # Social Media / Internet
    if any(word in text for word in ['ุณูุดูุงู', 'ููุฏูุง', 'ุงูุชุฑูุช', 'ููุงูุน ุงูุชูุงุตู', 'ูุณุงุฆู ุงูุชูุงุตู', 'social']):
        categories.append('Social Media / Internet')

    # Friends / Social Life
    if any(word in text for word in ['ุงุตุฏูุงุก', 'ุงูุฎุฑูุฌ', 'ุงูุณูุก']):
        categories.append('Friends / Social Life')

    # Games / Sports
    if any(word in text for word in ['ุงูุนุงุจ', 'ูุจุงุฑูุงุช']):
        categories.append('Games / Sports')

    # TV / Series
    if 'ูุณูุณูุงุช' in text:
        categories.append('TV / Series')

    # Family / Visits
    if any(word in text for word in ['ุงูุงุณุฑู', 'ุนุงุฆูุฉ', 'ุฒูุงุฑุงุช', 'ุงูุงุฑุจ']):
        categories.append('Family / Visits')

    # Lack of Motivation / Focus
    if any(word in text for word in ['ุญุงูุฒ', 'ูุฏู', 'dreaming']):
        categories.append('Lack of Motivation / Focus')

    if len(categories) == 0:
        categories.append('Other')

    return ', '.join(categories)
````

````python
df['study_distraction_clean'] = df['study_distraction'].apply(clean_study_distraction)
````

````python
categories = [
    'Phone / Screens',
    'Social Media / Internet',
    'Friends / Social Life',
    'Games / Sports',
    'TV / Series',
    'Family / Visits',
    'Lack of Motivation / Focus',
    'Other'
]

for cat in categories:
    df[cat] = df['study_distraction_clean'].apply(lambda x: 'Yes' if cat in str(x) else 'No')
````
